````python
import pyinputplus as pyip

name = pyip.inputStr(":\n")
height = pyip.inputFloat('(120~230)(cm):\n',min=120,max=230)
weight = pyip.inputFloat('(40~170)(kg):\n',min=40,max=170)

bmi = weight / (height / 100) ** 2
print(f'{name}, BMI: {round(bmi,ndigits=2)}')
````
