## 請使用自訂的function
* 建立bmi的function,會傳出bmi
* 建立get_status,會傳出字串
* 可以使用pypi

## 輸入:
* 請輸入姓名:xxxx
* 請輸入身高(cm):
* 請輸入體重(kg):

## 輸出:
* xxxx,您的BMI:25.52
* xxxx,您的體重過重
## 錯誤:
* 格式錯誤:xxxxxxxxxxxxx


```python
import pyinputplus as pyip




def BMI(height:float,weight:float):

    bmi = weight / (height / 100) ** 2

    return bmi

def get_status(bmi:float):
    if bmi < 18.5:
        result = '過輕'
    elif bmi < 24:
        result = '正常'
    elif bmi < 27:
        result = '過重'
    elif bmi < 30:
        result = '輕度肥胖'
    elif bmi < 35:
        result = '中度肥胖'
    else:
        result = '重度肥胖'
    return result



name:str = input('請輸入姓名: \n')
height:float = pyip.inputFloat('請輸入身高(120~230)(cm): \n', min=120, max=230)
weight:float = pyip.inputFloat('請輸入體重(40~170)(kg): \n', min=40, max=170)

bmi = BMI(height,weight)
status = get_status(bmi)
print(f'{name}, 您的BMI: {round(bmi,ndigits=2)}')
print(f'您的體重: {status}')
```
