```python
import pyinputplus as pyip




def BMI(height:float,weight:float):

    bmi = weight / (height / 100) ** 2

    return bmi

def get_status(bmi:float):
    if bmi < 18.5:
        result = ''
    elif bmi < 24:
        result = ''
    elif bmi < 27:
        result = ''
    elif bmi < 30:
        result = ''
    elif bmi < 35:
        result = ''
    else:
        result = ''
    return result



name:str = input(': \n')
height:float = pyip.inputFloat('(120~230)(cm): \n', min=120, max=230)
weight:float = pyip.inputFloat('(40~170)(kg): \n', min=40, max=170)

bmi = BMI(height,weight)
status = get_status(bmi)
print(f'{name}, BMI: {round(bmi,ndigits=2)}')
print(f': {status}')
```
