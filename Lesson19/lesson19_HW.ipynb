```python
import os
import pandas as pd

def make_data_frame(file_path: str) -> pd.DataFrame:
    df = pd.read_csv(file_path)
    df.columns = df.iloc[0]
    df1 = df.drop(0).dropna()
    df2 = df1.set_index('')
    return df2


dir_ = ''
dfs = []
for file in os.listdir(dir_):
    dfs.append(make_data_frame(f'{dir_}/{file}'))

pd.concat(dfs)
```
