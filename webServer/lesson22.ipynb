```python
import os
import google.generativeai as genai

# 設定 API 金鑰
genai.configure(api_key=os.environ['Gemini_API_KEY'])

# 使用新版模型名稱
model = genai.GenerativeModel(model_name='models/gemini-2.5-flash')

# 產生回應
response = model.generate_content('我想要學習演算法')
print(response.text)
```
